##**Business Objective**

Dataset Medical Insurance Cost

Analisis pada dataset Medical Insurance Cost ini bertujuan untuk membantu perusahaan asuransi memahami faktor-faktor yang mempengaruhi besarnya biaya asuransi. Dataset yang digunakan berisi data seperti usia, jenis kelamin, BMI, jumlah anak, kebiasaan merokok, dan wilayah tempat tinggal.

Dengan melakukan analisis ini, perusahaan diharapkan medapat insight mengenai variabel mana yang memberikan pengaruh signifikan terhadap peningkatan biaya. Hasil analisis ini dapat digunakan untuk memdukung pengambilan keputusan dalam penetapan premi yang lebih akurat dan berbasis resiko.

#**Asses Situation**

Situasi bisnis yang mendasari analisis ini adalah meningkatnya biaya layanan kesehatan yang berdampak langsung terhadap besarnya klaim asuransi. Perusahaan asuransi perlu memastikan bahwa premi yang ditetapkan sesuai dengan tingkat resiko masing-masing pelanggan. Jika premi terlalu rendah, perusahaan bisa mengalami kerugian. Jika terlalu tinggi, pelanggan bisa merasa dirugikan.

Dalam industri asuransi kesehatan, setiap pelanggan memiliki tingkat risiko yang berebeda-beda. Risiko ini bisa dipengaruhi oleh usia, BMI, jumlah tanggungan anak, serta kebiasaan merokok. Tanpa analisis data yang tepat perusahaan akan sulit menentukan premi yang sesuai dengan tingkat risiko masing-masing pelanggan.
